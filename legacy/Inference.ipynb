```python
from modules.trainers.DDPM import DDPM
ddpm = DDPM.load_from_checkpoint('./genmod/ixicduzq/checkpoints/last.ckpt')
ddpm.eval()

device = "cuda:0"
ddpm = ddpm.to(device)

with torch.no_grad():
    x_i = torch.randn((9,1,28,28)).to(device)  # x_T ~ N(0, 1), sample initial noise
    x_0 = ddpm.reverse_diffusion_process(x_i)

grid = make_grid(x_0.cpu(), nrow=3, padding=2)

plt.imshow(grid.permute(1, 2, 0))
plt.axis('off')
plt.show()
```
